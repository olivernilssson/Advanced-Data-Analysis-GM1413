### Calculating Market Ratios
Objective: Use the generated financial data from Exercise 1 to calculate several market ratios.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

print(f"NumPy version: {np.__version__}")
print(f"Pandas version: {pd.__version__}")
print(f"Matplotlib version: {plt.matplotlib.__version__}")
```

```python
# Import Data from CSV from previous exercise
df_combined = pd.read_csv("../data/company_data.csv")

# Calculate EPS
df_combined["EPS"] = df_combined["Total Earnings"] / df_combined["Number of Shares Outstanding"]

# Calculate PE Ratio
df_combined["PE"] = df_combined["Stock Price"] / df_combined["EPS"]

# Calculate Market Capitalization
df_combined["Market Cap"] = df_combined["Stock Price"] * df_combined["Number of Shares Outstanding"]

# Calculate Dividend Yield
df_combined["Dividend %"] = df_combined["Dividend"] / df_combined["Stock Price"] * 100

# Display Head
df_combined.head()
```

```python
# PLot EPS over time for Company A and Company B
plt.figure(figsize=(10, 6))
plt.plot(range(0, len(df_combined.loc[df_combined["Company"] == "A"])), df_combined["EPS"].loc[df_combined["Company"] == "A"], label="Company A")
plt.plot(range(0, len(df_combined.loc[df_combined["Company"] == "B"])), df_combined["EPS"].loc[df_combined["Company"] == "B"], label="Company B")
plt.xlabel("Day")
plt.ylabel("EPS")
plt.legend()
plt.grid(True)
plt.title("EPS over time for Company A and Company B")
plt.show()
```
