### Generating Fundamental Financial Data
Objective: Generate realistic financial data for two companies, focusing on fundamental data points needed to calculate key financial ratios in the next exercise.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

print(f"NumPy version: {np.__version__}")
print(f"Pandas version: {pd.__version__}")
print(f"Matplotlib version: {plt.matplotlib.__version__}")
```

```python
# Set the random seed for reproducibility
np.random.seed(10)

# Number of data points
n = 100

def generate_company_data(company_name, n, stock_price_range, total_earnings_range, shares_outstanding_range, 
                          total_debt_range, total_equity_range, net_income_range, dividend_range):
    stock_price = np.random.uniform(*stock_price_range, n)
    total_earnings = np.random.uniform(*total_earnings_range, n)
    shares_outstanding = np.random.uniform(*shares_outstanding_range, n)
    total_debt = np.random.uniform(*total_debt_range, n)
    total_equity = np.random.uniform(*total_equity_range, n)
    net_income = np.random.uniform(*net_income_range, n)
    total_assets = total_debt + total_equity
    dividend = np.random.uniform(*dividend_range, n)
    
    return {
        "Company": company_name,
        "Stock Price": stock_price,
        "Total Earnings": total_earnings,
        "Number of Shares Outstanding": shares_outstanding,
        "Total Debt": total_debt,
        "Total Equity": total_equity,
        "Net Income": net_income,
        "Total Assets": total_assets,
        "Dividend": dividend
    }

# Company A Data
df_a = pd.DataFrame(generate_company_data(
    "A",
    n,
    stock_price_range=(10, 100),
    total_earnings_range=(1e6, 1e7),
    shares_outstanding_range=(10e6, 70e6),
    total_debt_range=(6e5, 2e6),
    total_equity_range=(1e6, 3e6),
    net_income_range=(5e5, 3e6),
    dividend_range=(0.1, 1.5)
))

# Company B Data
df_b = pd.DataFrame(generate_company_data(
    "B",
    n,
    stock_price_range=(15, 120),
    total_earnings_range=(2e6, 8e6),
    shares_outstanding_range=(8e6, 60e6),
    total_debt_range=(1e6, 3e6),
    total_equity_range=(1.5e6, 4e6),
    net_income_range=(6e5, 2.5e6),
    dividend_range=(0.05, 2.0)
))

df_a
```

```python
# Display Company B DataFrame
df_b
```

```python
# Use concat to vertically combine the two DataFrames
df_combined = pd.concat([df_a, df_b], ignore_index=True, axis=0)

# Display the combined DataFrame
df_combined
```

```python
# Display summary statistics for the combined DataFrame
df_combined.describe()
```

```python
# Correlation between Net Income and Total Assets to see if there are any patterns
df_combined[["Net Income", "Total Assets"]].corr()
```

```python
# Plot the data
plt.figure(figsize=(12, 6))
plt.subplot(2, 1, 1) # (rows, columns, index)
plt.plot(df_combined[df_combined["Company"] == "A"].reset_index(drop=True)["Stock Price"], color="blue", label="Company A")
plt.title("Company A Stock Prices")
plt.xlabel("Day")
plt.ylabel("Stock Price")
plt.legend()

plt.subplot(2, 1, 2)
plt.plot(df_combined[df_combined["Company"] == "B"].reset_index(drop=True)["Stock Price"], color="orange", label="Company B")
plt.title("Company B Stock Prices")
plt.xlabel("Day")
plt.ylabel("Stock Price")
plt.legend()

plt.tight_layout()
plt.show()
```

### More realistic price trend
Challenge: Instead of generating completely random stock prices, create a price trend with slight daily growth or decline and random fluctuations to simulate real-world price movements.

```python
# Set the random seed for reproducibility
np.random.seed(10)

# Number of data points
n = 100

initial_price = 50
daily_growth_rate = 0.001 # 0.1%
daily_volatility = 0.02 # 2%

stock_prices = [initial_price] # Add the initial price to the list

# Genereate stcok prices with a trend and volatility
for _ in range(1, n):
    price = stock_prices[-1]
    price *= 1 + daily_growth_rate + np.random.normal(0, daily_volatility)
    stock_prices.append(price)

# Create DataFrame for days and stock prices
df_stock_prices = pd.DataFrame({
    "Day": range(n),
    "Stock Price": stock_prices
})

# Plot the stock prices
plt.figure(figsize=(10, 6))
plt.plot(df_stock_prices["Day"], df_stock_prices["Stock Price"], color="green", label="Stock Price")
plt.title("Stock Price Over Time")
plt.xlabel("Day")
plt.ylabel("Stock Price")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Compare the random of Company A and realisitc stock prices
# Ensure the length is the same
assert len(df_a) == len(df_stock_prices)

# Plot the stock prices
plt.figure(figsize=(10, 6))
plt.plot(df_stock_prices["Day"], df_stock_prices["Stock Price"], color="green", label="Realistic Stock Price")
plt.plot(df_a["Stock Price"], color="blue", label="Company A Stock Price")
plt.title("Comparison of Stock Prices")
plt.xlabel("Day")
plt.ylabel("Stock Price")
plt.legend()
plt.grid(True)
plt.show()
```
