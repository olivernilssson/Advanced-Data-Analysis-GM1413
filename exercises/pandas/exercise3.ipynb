### Scaling Financial Data Using Pandas
Objective: Understand why scaling is necessary and apply scaling techniques to normalize the data using pandas.

In financial analysis, the data often includes variables with vastly different ranges (e.g., stock prices vs. earnings per share). Scaling ensures that each variable is normalized, preventing any one feature from dominating the analysis. Take the dataset from the previous exercise and scale the numerical columns such as Stock Price, Market Capitalization, and EPS to a 0-1 range using pandas.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

print(f"NumPy version: {np.__version__}")
print(f"Pandas version: {pd.__version__}")
print(f"Matplotlib version: {plt.matplotlib.__version__}")
```

```python
# Import Data from CSV from previous exercise
df_combined = pd.read_csv("../data/company_data.csv")

X_min = df_combined["EPS"].min()
X_max = df_combined["EPS"].max()
df_combined["EPS_scaled"] = (df_combined["EPS"] - X_min) / (X_max - X_min)
df_combined[["EPS", "EPS_scaled"]]
```

```python
# Plotting 'EPS' and 'EPS_scaled' in separate subplots
plt.figure(figsize=(12, 6))

# Subplot 1: Original EPS
plt.subplot(2, 1, 1)
plt.plot(df_combined.index, df_combined['EPS'], label='EPS', color='blue')
plt.xlabel('Index')
plt.ylabel('EPS Value')
plt.title('EPS')
plt.legend()
plt.grid(True)

# Subplot 2: Scaled EPS
plt.subplot(2, 1, 2)
plt.plot(df_combined.index, df_combined['EPS_scaled'], label='EPS Scaled', color='red')
plt.xlabel('Index')
plt.ylabel('Scaled EPS Value')
plt.title('Scaled EPS')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```
