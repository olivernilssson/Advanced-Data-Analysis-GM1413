# Exercise: Analysing stock data using NumPy

### Stage 2: Calculate Portfolio Returns Using Dot Product
Objectives: Given a set of stock weights in a portfolio, calculate the portfolio returns for each day using the dot product of stock prices and weights.

```python
import numpy as np
import matplotlib.pyplot as plt
import datetime

print(f"NumPy version: {np.__version__}")
print(f"Matplotlib version: {plt.matplotlib.__version__}")
print(f"Date: {datetime.datetime.now()}")
```

```python
# Load Data from CSV
dates = np.genfromtxt("../data/numpy1.csv", delimiter=",", dtype="str", skip_header=1, usecols=0)
prices = np.genfromtxt("../data/numpy1.csv", delimiter=",", dtype="float", skip_header=1, usecols=(1, 2, 3))

dates = [datetime.datetime.strptime(date, "%Y-%m-%d") for date in dates]
```

```python
# Define weights for the three stocks in the portfolio
weights = np.array([0.4, 0.3, 0.3])

# Ensure weights sum to 1
assert np.sum(weights) == 1

# Ensure the shape of the weights and prices array is compatible
assert prices.shape[1] == weights.shape[0]

# Calculate the portfolio value for each day using a dot product
portfolio_values = np.dot(prices, weights)

# Print the portfolio value for each day
print("Portfolio Value for Each Day:")
for date, p_value in zip(dates, portfolio_values):
    print(f"{date.date()}: {p_value:.2f}")
```

```python
# Plot the portfolio value over time
plt.figure(figsize=(12, 6))
plt.plot(dates, portfolio_values, marker="o", label="Portfolio Value", color="blue")
plt.xlabel("Date")
plt.ylabel("Portfolio Value")
plt.title("Portfolio Value Over Time")
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout
plt.show()
```
